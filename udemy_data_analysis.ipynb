<b><h5>10. What are the Max. Number of subscribers for Each level of Courses ?<h><b>

```python
df.level.unique()	
```

```python
df.groupby('level')['num_subscribers'].max()
```

```python
max_sub_by_level=df.groupby('level')['num_subscribers'].max().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=max_sub_by_level,x="level",y="num_subscribers")
plt.title("Maximum Number of subscribers by Course Level")
plt.xlabel("Course Level")
plt.ylabel("Max Number of Subscribers")
plt.show()
```
